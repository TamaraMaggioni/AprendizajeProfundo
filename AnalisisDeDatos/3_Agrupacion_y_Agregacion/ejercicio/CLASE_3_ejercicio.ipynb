```python
df = df.dropna(axis=0).copy()

df.sort_values('Codigo')
```

Unan ambos DataFrames utilizando la función pd.merge()

```python
df_mergeado = df.merge(
    df_descripciones
)

df_mergeado.head()
```

```python
df_mergeado.Texto.nunique(), df_mergeado.Codigo.nunique()
```

```python
df_mergeado[df_mergeado.isna().any(axis=1)]
```

```python
df_mergeado = df_mergeado.rename(columns=
                                  {'Texto' : 'Municipios'})
```

Construir el DataFrame de comunidades autónomas, incluyendo la densidad, la proporción de paro y la cantidad de municipios. 

```python
df_mergeado['Densidad'] = df_mergeado['PAD_1C02'] / df_mergeado['Shape__Area']
df_mergeado['Proporcion_Paro'] = df_mergeado['TotalParoRegistrado'] / df_mergeado['PAD_1C02']
```

```python
comunidades = df_mergeado.groupby('Comunidad_Autónoma').aggregate({
    'Densidad' : 'mean',
    'Proporcion_Paro' : 'mean',
    'Municipios' : 'count'
}).reset_index()

comunidades.head()
```

Habíamos visto que el promedio de la proporción de paro en los municipios con una densidad menor a la mediana es más bajo. ¿Se cumple lo mismo para las ciudades autónomas?

```python
com_media = comunidades.Densidad.median()
```

```python
comunidades.query('Densidad > @com_media')['Proporcion_Paro'].mean()
```

```python
comunidades.query('Densidad <= @com_media')['Proporcion_Paro'].mean()
```

**Comunidades autónomas con densidades más altas a la media tienen mayor tasa de desempleo que las comunidades autónomas con densidades menores a la media.**

¿Cuál es la comunidad autónoma con mayor cantidad de municipios?

```python
comunidades[comunidades.Municipios == comunidades.Municipios.max()]
```
